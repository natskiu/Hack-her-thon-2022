```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python

df = pd.read_excel (r'/Users/jiyeonkim/Documents/food_data_simple.xlsx',header=0)
df
```

```python
df.columns
df.columns = df.columns.str.lstrip()
df.set_index('Food commodity ITEM')
df
```

```python
typologies = df['Food commodity TYPOLOGY'].unique()
```

```python

x= df['Food commodity TYPOLOGY']
for f in typologies:
    print(f)
    food_type = f
    y_category = []
    x_category = []
    count = -1
    for i in x: 
        count += 1
        if i == food_type:
            y_category.append(df['Carbon Footprint kg CO2eq/kg or l of food ITEM'].iloc[count])
            x_category.append(df['Food commodity ITEM'].iloc[count])
        
    plt.figure(figsize=(9,6))

    plt.bar(x=x_category,height=y_category)
    plt.xticks(rotation=45)

   # print(y_category)
    #print(x_category)
    
```
